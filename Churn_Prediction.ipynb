```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/drive/My Drive/Churn_Modelling.csv')
df
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df = pd.get_dummies(df, columns = ["Geography","Gender"], drop_first=True)
```

```python
df
```

```python
x = df.drop("Exited", axis=1).iloc[:,3:]
y = df["Exited"]
```

```python
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()
x_scalar = pd.DataFrame(scalar.fit_transform(x), columns=x.columns)
```

```python
x_scalar.shape[1]
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x_scalar, y, test_size=0.2, random_state=10)
```

```python
xtrain
```

```python
ytrain
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
```

```python
ann_model=tf.keras.models.Sequential()
ann_model
```

```python
ann_model.add(Dense(units=3, activation='sigmoid',input_dim=11)) # 1st hiddin layer
```

```python
ann_model.add(Dense(units=3, activation='sigmoid', input_dim=11)) # 2nd hidden layer
```

```python
ann_model.add(Dense(units=1, activation='sigmoid')) #o/p layer
```

```python
ann_model.summary()
```

```python
ann_model.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy', 'precision', 'recall'])
```

```python
ann_model.fit(xtrain, ytrain, epochs=100, batch_size=32, validation_split=0.2)
```

```python
x.columns
```

```python
if ann_model.predict(scalar.transform([[700, 35, 1, 80000, 2, 1, 1, 90000, 1, 0, 1]])) > 0.5 == 1:
  print("Customer will leave the bank")
else:
  print("Customer will stay")
```

```python
ytest_pred = ann_model.predict(xtest)
ytest_pred = np.where(ytest_pred > 0.5, 1,0)
ytest_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
CM = confusion_matrix(ytest, ytest_pred)
print("\n",CM,"\n")

accuracy_score(ytest, ytest_pred)
```
